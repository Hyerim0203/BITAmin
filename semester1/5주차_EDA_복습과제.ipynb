# 문제3
### PatientRoute.csv 데이터를 활용하여 확진자가 이동한 장소 중 서울역까지의 거리가 가장 짧은 장소와 서울역 사이의 거리를 확진자별로 구해주세요.
- latitude(위도), longitude(경도)를 이용하여 거리를 계산해주세요<br/>
  (서울역의 위도 = 37.5447034, 경도 = 126.9451916 입니다)
- 아래와 같은 내용이 출력되면 됩니다.(형식은 상관없습니다)

   patient_id|   |
    :-:|:-:|
    1000000002|7.740634|
    ...|...|
    ...|...|
    ...|...|
    1000000434|7.510743|
    

```python
from math import sin, cos, sqrt, atan2, radians
R = 6373.0
# mydistance 함수는 위도 경도를 통해 특정 장소에서 서울역까지의 거리를 계산해주는 함수입니다.
# 함수의 input값 a1,b1은 특정 장소의 경도, 위도 값이 되겠죠?
# 구하는 과정은 모르셔도 상관없습니다! 저도 잘 모릅니다..
def mydistance(a1,b1):
    x1 = radians(a1)
    y1 = radians(b1)
    x2 = radians(126.9451916)
    y2 = radians(37.5447034)
    dlon = abs(x2 - x1)
    dlat = abs(y2 - y1)
    a = sin(dlat / 2)**2 + cos(y1) * cos(y2) * sin(dlon / 2)**2 
    c = 2 * atan2(sqrt(a), sqrt(1 - a))    
    return R * c
```

```python
seoul_df=patient_df.loc[patient_df["province"]=="Seoul",:]
seoul_df.merge(case_df,on=["patient_id"],how="left")
seoul_df.apply()
```

```python
#위의 mydistance 함수를 사용하여 'distance' 열을 생성합니다.
route_df['distance']= route_df.apply(lambda x: mydistance(x['longitude'],x['latitude']),axis=1)
```

```python
route_df
```

```python
#patient_id를 기준으로 그룹화한 후 확진자별 distance의 최솟값을 구해주시면 됩니다.
route_df.groupby("patient_id")["distance"].agg("mean")
```

# 문제4
### PatientInfo.csv 데이터를 활용하여 성별, 연령대별 사망자수를 구해주세요.
- 그룹의 단순화를 위해 birth_year를 기준으로 1990년대 이상이면 '청년층', 1970~1990 이면 '중장년층',<br/>
  1970 이하면 '노년층' 으로 범주화 하겠습니다.
- 사망자의 경우 state 열에 'deceased'로 표시되어 있습니다.
- 성별(male,female), 연령대(청년층, 중장년층, 노년층)로 그룹화 한 후 그룹별 사망자수를 출력하시면 됩니다 
- hint)수업시간에 배운대로 범주화 함수를 만든 후 apply lambda를 활용해주세요 (지난주에 배운 pd.cut 함수 이용해도 됩니다)

  아래와 같은 내용이 출력되면 됩니다.

   sex|age_cat|  |
    :-:|:-:|:-:|
    female|노년층|16
       |중장년층|0
       |청년층|0
    male|노년층|32
       |중장년층|2
       |청년층|0

```python
def birth(x):
    age_list = ["청년층","중장년층","노년층"]
    if x>=1990:
        return age_list[0]
    elif x>=1970 and x<1990:
        return age_list[1]
    elif x<1970:
        return age_list[2]
```

```python
# 데이터 중 sex, birth_year, state 값이 없는 데이터 제거 후 새로운 데이터프레임 생성
new_patient_df = patient_df.dropna(subset=['sex','birth_year','state'])
```

```python
new_patient_df["age_cat"]=patient_df.apply(lambda x: birth(x["birth_year"]), axis=1)
```

```python
new_patient_group=new_patient_df.groupby(["sex","age_cat"])
```

```python
new_patient_group.agg({"state":lambda x: sum(x=="deceased")})
```
