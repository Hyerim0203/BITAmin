folium.Choropleth를 활용하여 location=\[48, -102\], zoom_start=3 인 지도 위에 각 주마다 Covid-19 진료 지점 수를 단계 구분도로 나타내 Bitamin_week6_Homwork6_Result_02.html 과 같은 파일을 만드세요.<br>
<br>
1) 'time_series_covid_19_confirmed_US.csv'를 읽어오세요.<br>
2) 'Province_State'의 값들을 카운트해 새로운 시리즈(or 데이터프레임)을 만드세요.<br> 
3) 'Province_State' 중 실제 미국 주가 아닌 것들을 제거하세요.<br>
( 위의 'us-state-capitals.csv' 파일 활용)<br>
3) 'us-states.json'을 읽어오세요.<br>
4) 아래와 같은 옵션으로 단계구분도를 만드세요.<br>
fill_color='YlOrRd'<br>
fill_opacity=0.7,<br>
line_opacity=0.3,<br>
threshold_scale=[0, 20, 50, 100, 500]<br>

```python
covid=pd.read_csv("./time_series_covid_19_confirmed_US.csv")
```

```python
count = pd.DataFrame(covid["Province_State"].value_counts())
count["name"]=count.index
```

```python
capital= pd.read_csv("./us-state-capitals.csv")
new_count = capital.merge(count,on=["name"],how="left")
new_count.index=new_count["name"]
```

```python
import json
maped2 = folium.Map(location=[48,-102], zoom_start=3 )
geo_path="./us-states.json"
try:
    geo_data = json.load(open(geo_path,encoding="utf-8"))
except:
    geo_data = json.load(open(geo_path,encoding="utf-8-sig"))

folium.Choropleth(geo_data=geo_data, data=new_count["Province_State"],columns=[new_count.index,new_count["Province_State"]], line_opacity=0.3,fill_opacity=0.7,
    fill_color="YlOrRd",threshold_scale=[0,20,50,100,500],key_on="feature.properties.name").add_to(maped2)
maped2.save("haha.html")
```
