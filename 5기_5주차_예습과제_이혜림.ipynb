## 데이터프레임 단순히 연결
- concat() : 
    - 행방향(default) -> ignore_index=False 
    - 열방향
    - 시리즈&시리즈, 시리즈&df, df&df 모두 가능


```python
import pandas as pd

df1=pd.DataFrame({"a":"a0 a1 a2 a3".split(),"b":"b0 b1 b2 b3".split(),"c":"c0 c1 c2 c3".split()},
                index=[0,1,3,2])
df2=pd.DataFrame({"a":"a2 a3 a4 a5".split(),"b":"b2 b3 b4 b5".split(), "c":"c2 c3 c4 c5".split(),
                "d":"d2 d3 d4 d5".split()}, index=[2,3,4,5])
print(df1,"\n",df2)
```

```python
print(pd.concat([df1,df2])) # 없는 열에 대해서는 nan 값 반환
print(pd.concat([df1,df2],ignore_index=True)) # 두 DP에 대해 모두 새로운 인덱스 설정.
print(df1.append(df2,ignore_index=True))
print(df1.append(df2,ignore_index=True))
```

```python
print(pd.concat([df1,df2],axis=1, join="inner", sort=False)) #인덱스를 sort 해줌
print(pd.concat([df1,df2],axis=1, join="outer"))
```

```python
# 시리즈 만들기
sr1 = pd.Series(["e0","e1","e2","e4"],name="e")
sr2 = pd.Series(["f0","f1","f2"],name="f", index=[1,5,3])
```

```python
print(pd.concat([sr1,df1],axis=1))
print(pd.concat([sr2,df1],axis=1))
print(pd.concat([sr1,df1]))
```

## 데이터프레임 병합
- merge 
    - on=None(default) : 두 df 에 속하는 모든 열을 키로 하여 병합
    - how  
        -"inner"(default) : key의 데이터가 두 df 에 공통으로 존재하는 교집합인 경우에만 추출
        - "outer" : 한 df에만 속하더라도 추출, 합집합
        - "left" / "right"
- join : 행 인덱스를 기준으로 병합. how="inner"
- append : 열 인덱스를 기준으로 병합.

## 그룹연산
- grouped
    - :'class' 열 기준으로 그룹화된 데이터프레임 객체
    - : 분할만 이루어진 상태이기 때문에 데이터 출력 결과 확인 불가능
        - 연산 메소드/get_group()메소드 적용시 출력 가능
            - 선택적 연산 수행 : 연산이 불가능한 문자열은 알아서 제외(max/min 안됨)
    - key: 그룹 객체에서 기준이 되는 열의 분류 클래스
    - group : 클래스 별로 분류된 데이터 집합(각 행의 행 인덱스 유지)
    - get_group()
    - agg : 각 칼럼 별로 함수 적용시 클래스 별로 분류된 형태로 값을 반환
    - transform : 각 칼럼 별로 함수 적용시 원본 df 의 형태로 값을 반환
    - apply : 각 그룹 별로 함수 매핑, 클래스 별로 분류된 형태로 값 반환
    - filter : 각 그룹에 대해 조건을 만족하는 그룹만을 선택하여 transform 의 결과처럼 원본 데이터프레임 형식으로 결과 반환

```python
# 한개의 칼럼 기준 groupby
for key, group in titanic.groupby(["class"]):
    print(f"*key : {key}")   # key : 그룹 객체에서 기준이 되는 열의 분류 클래스
    print(f"*number : {len(group)}") 
    print(group.head()) # group : 클래스별로 분류된 데이터 집합
    print("")
group=titanic.groupby(["class"])  
print(group.std())
print(group[["age","survived"]].mean())
print(group.get_group("Third"))
```

```python
# 두개의 칼럼 기준 groupby -> 기준이 되는 열이 가지는 모든 값의 조합으로 생성
group2 = titanic.groupby(["class","sex"])
for key, group in group2:
    print(f"*key : {key}")
    print(f"*number : {len(group)}")
    print(group.head())

group2.mean()
group2.get_group(("First","female")) #튜플로 접근해야 함
```

```python
# 집계 함수 전체, 부분 적용
display(group.std())
display(group[["alone","pclass"]].std())

# 모든 열에 대해 하나의 함수 적용
def min_max(x):
    return x.max()-x.min()

def describe(x): # series에는 적용 불가능, 따라서 열별로 함수를 매핑하는 agg,transform 사용 X
    return x.describe()
display(group.agg(min_max))
display(group.apply(lambda x: x.describe()))

#선택적 열에 대해 여러 함수 적용, 여러개의 층 열 구조가 반환
display(group[["pclass","survived"]].agg([min_max,max,min]))
#각 열에 다른 함수를 적용하는 경우
display(group.agg({"fare":min_max,"age":[min,max]}))
```

```python
age_mean = group.age.mean()
age_std = group.age.std()

for key, g in group.age:
    a= (g-age_mean.loc[key])/age_std.loc[key]
    print(a.head(3))

# for 문을 돌면서 추출된 data의 key 에 맞는 아이들을 빼주고 더해줌!
```

### transform

```python
def z_score(x):
    return (x-x.mean())/x.std()

t_group=group.transform(z_score)
# 함수 적용시, group df의 형태가 아닌 일반 df 형태로 값을 반환해줌
t_group.head(5)
```

### filter

```python
display(group.filter(lambda x:len(x)>=400))

display(group.apply(lambda x: x.age.mean()<30)) # filter와 달리 class 별로 불리언 값
age_filter = group.apply(lambda x:x.age.mean()<30)
for i in age_filter.index:
    if age_filter[i]:
        age_filter_df=group.get_group(i)
        print(age_filter_df)
```

## 멀티 인덱스(이거와 이거x -> 이거에서의 이거) () 형태로 써주어야 함!
1. loc 인덱서 -> 여러 열을 선택할 때 맨 상위 레벨에서부터 차례차례 순서에 맞게 튜플 형태로 전달
1. xs 인덱서 -> xs(인덱스,level) -> level만 기입해주면 순서 틀려도, 상위부터 안타고 내려와도 됨


```python
grouped = titanic.groupby(["class","sex"])
gdf = grouped.mean()
display(gdf)

display(gdf.loc["First"])
display(gdf.loc[("First","female")]) #[]을 하게 되면 "First" 행과 "female" 행을 가져와라. 따라서 ()로 묶어주어야 함!

display(gdf.xs("male", level="sex"))
display(gdf.xs(["female","Second"],level=["sex","class"]))
```

## 피벗테이블
- 판다스객체.pivot_table(데이터프레임,index=‘ ‘, columns=‘ ‘, values=‘ ‘, aggfunc=‘ ‘)


```python
pdf1 = pd.pivot_table(titanic,
                     index="class",
                     columns="sex",
                     values="age",
                     aggfunc = ["mean"]) # 예외적으로 함수에 "" 붙임

display(pdf1)
```

```python
pdf2 = pd.pivot_table(titanic,
                     index=["class","sex"],
                     columns="survived",
                     values=["age","fare"],
                     aggfunc=["mean","max"])
display(pdf2)
print(pdf2.index)
print(pdf2.columns)
```

```python
# 행선택
pdf2.xs("female",level=1)
pdf2.xs(["female","First"],level=[1,0])
pdf2.xs(["female","First"],level=[1,"class"])
```

```python
# 열선택
display(pdf2.xs("mean",axis=1))
display(pdf2.xs(("age","max"),axis=1,level=[1,0]))
display(pdf2.xs(0,axis=1,level="survived"))
display(pdf2.xs(0,axis=1,level=2))
```
